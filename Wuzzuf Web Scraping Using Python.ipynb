```python
f = open("D:\Data Analyst Project\Data_webscraping\Wuzzuf-illustrator.csv" , "w")
header = "job_title, company_Name, job_Type\n"
f.write(header)
```

```python
 for container in containers :
    Job_title =container.findAll("h2",{"class" : "css-m604qf"})
    job_title = Job_title[0].text.strip()
    
    cname = container.findAll("a",{"class" : "css-17s97q8"})
    company_Name=cname[0].text.strip()
    
    Jobtype = container.findAll("span",{"class" : "css-1ve4b75 eoyjyou0"})
    job_Type = Jobtype[0].text.strip()
    
    #print(job_title)
    #print(company_Name)
    #print(job_Type)
    #print()
    
    #print(job_title + " , " + company_Name +" , " + job_Type)
    #print()
f.write(job_title + " , " + company_Name +" , " + job_Type +"\n")  
f.close()
```
